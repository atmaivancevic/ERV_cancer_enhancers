```python
import plotly.express as px
import pandas as pd

df = pd.read_csv('gnomad.indels.10to300bp.tab', sep='\t')
#print(df)

# make a scale so that the bubbles are more visible

all_data_diffq = (df["af"].max() - df["af"].min()) / 30
df["scale"] = (df["af"] - df["af"].min()) / all_data_diffq + 1

fig = px.scatter(df, x="length", y="subfamily", size=df["scale"],
                title="Show all subfamilies, with bubble size as allele frequency")

for template in ["plotly_white"]:    
    fig.update_yaxes(categoryorder="category descending")
    fig.update_layout(template=template)
    #fig.update_layout(autosize=False,width=2000,height=1000)
    
#fig.update_traces(quartilemethod="linear", jitter=1)
    
fig.update_xaxes(title_standoff=0)
fig.update_yaxes(title_standoff=5)
fig.update_xaxes(tickfont_size=18, ticks="outside", ticklen=5, tickwidth=2)
fig.update_yaxes(tickfont_size=18, ticks="outside", ticklen=5, tickwidth=2)

fig.update_xaxes(showline=True, linewidth=2.5, linecolor='black', gridwidth=1, tickangle=45)
fig.update_yaxes(showline=True, linewidth=2.5, linecolor='black', gridwidth=1)

#fig.update_layout(
#    font_family="Helvetica",
#    font_color="black",
#    font_size=16,
#)

fig.update_yaxes(nticks=50, tickangle=0, title_standoff=20)
fig.update_xaxes(nticks=20, tickangle=0)
#fig.update_layout(yaxis_range=[0,0.85])
fig.update_layout(xaxis_range=[8,150])

fig.update_layout(
    xaxis_title_text='Length of indel (bp)', # xaxis label
    yaxis_title_text='Subfamily', # yaxis label
)

fig.layout.font.family = 'Helvetica'
fig.update_layout(font=dict(size=16))

fig.show()
#fig.write_image("gnomad_swarm_plot.svg")
```
